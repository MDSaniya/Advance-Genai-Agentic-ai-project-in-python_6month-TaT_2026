```python
import sys
val1 = 10 # int data type
print(val1)
print(type(val1)) #type of obj
print(sys.getsizeof(val1)) # size of int obj in bytes
print(val1, "is Integer?", isinstance(val1, int)) # val1 is an instance of int
```

```python
import sys
val2 = 92.67
print(val2)
print(type(val2))
print(sys.getsizeof(val2))
print(val2, " is float?", isinstance(val2, float))
```

```python
import sys
val3 = 25 + 10j # complex data type
print(val3)
print(type(val3))
print(sys.getsizeof(val3))
print(val3, "is complex?", isinstance(val3, complex))
```
