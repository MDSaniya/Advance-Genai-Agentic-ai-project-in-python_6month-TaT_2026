```python
# rithmetic
z1 + 2
```

```python
# Subtraction
z1 - 2
```

```python
# Multiplication
z1 * 2
```

```python
# power
z1 ** 2
```

```python
# modulo
z1 % 2
```

## Array Functions

```python
k1 = np.random.random((3,3))
k1 = np.round(k1*100)
k1
```

```python
# Max
np.max(k1)
```

```python
# Min
np.min(k1)
```

```python
# sum
np.sum(k1)
```

```python
# prod -----> Multiplication
np.prod(k1)
```

## In Numpy
#### 0 = column, 1 = row


```python
# if we want maximum of every row
np.max(k1, axis = 1)
```

```python
# maximum of every column
np.max(k1, axis = 0)
```

```python
# product of evry column
np.prod(k1, axis = 0)
```
