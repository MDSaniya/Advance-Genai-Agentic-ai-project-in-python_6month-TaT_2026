## Slicing
##### Fetching Multiple items

### Slicing on 1D

```python
p1
```

```python
#  fetching desired element : 2,3,4
p1[2:5]
# Here First we take , whatever we need first item ,2 and up last(4) + 1 which 5 
#.because last element is not included
```

```python
#  Alternate (same as python) 
p1[2:5:2]
```

### Slicing on 2D

```python
p2
```

```python
# fetching total first row
p2[0,:]
# Here 0 represents first row and (:) represnts Total column
```

```python
# fetching total third column
p2[:,2]
# Here we want all rows so (:) , and we want 3rd column so 2 
```

```python
# fetch 5,6 and 9,10
p2
```

```python
p2[1:3] # for rows
```

```python
p2[1:3 , 1:3] # for columns
# Here first [1:3] we slice 2 second row is to third row is not existed which is 2 
# and Secondly , we take [1:3] which is same as first:we slice 2 second row is to third row is not 
# included which is 3
```

```python
# fetch 0,3 and 8, 11
p2
```

```python
p2[::2, ::3]
# Here we take (:) because we want all rows , second(:2) for alternate value, 
# and (:) for all columns and (:3) jump for two steps
```

```python
# fetch 1,3 and 9,11
p2
```

```python
p2[::2] # for rows
```

```python
p2[::2 , 1::2] # columns
# Here we take (:) because we want all rows , second(:2) for alternate value, 
# and (1) for we want from second column and (:2) jump for two steps and ignore middle one 
```

```python
# fetch only 4,7
p2
```

```python
p2[1] # first rows
```

```python
p2[1,::3] # second columns
# Here we take (1) because we want second row , second(:) for total column, 
# (:3) jump for two steps and ignore middle ones
```

```python
# fetch 1,2,3 and 5,6,7
p2
```

```python
p2[0:2] # first fetched rows
```

```python
p2[0:2 , 1:] # for column
```

```python
# fetch 1,3 and 5,7
p2
```

```python
p2[0:2] # for rows
```

```python
p2[0:2 , 1::2]
#  0:2 selects the rows from index 0 (inclusive) to index 2 (exclusive), which 
# means it will select the first and second rows of the array. , is used to separate row and column 
# selections. 1::2 selects the columns starting from index 1 and selects every second column. So 
# it will select the second and fourth columns of the array. 
```

### Slicing in 3D

```python
p3 = np.arange(27).reshape(3,3,3)
p3
```

```python
# fetch second matrix
p3[1]
```

```python
# fetch first and last
p3[::2]
#  Along the first axis, (::2) selects every second element. This means it will 
# select the subarrays at indices 0 and 2
```

```python
# Fetch 1 2d array's 2 row ---> 3,4,5 
p3
```

```python
p3[0] # first numpy array
```

```python
p3[0,1,:]
# 0 represnts first matrix , 1 represents second row , (:) means total 
```

```python
# Fetch 2 numpy array ,middle column ---> 10,13,16
p3
```

```python
p3[1] # middle Array
```

```python
p3[1,:,1]
# 1 respresnts middle column , (:) all columns , 1 represnts middle column
```

```python
#  Fetch 3 array--->22,23,25,26
p3
```

```python
p3[2] # last row
```

```python
p3[2,1:] # last two rows
```

```python
p3[2,1:,1:] # last two columns
#Here we go through 3 stages , where 2 for last array , and (1:) from second 
# row to total rows , and (1:) is for second column to total columns
```

```python
# fetch 0, 2, 8, 20
p3
```

```python
p3[0::2] # for arrays
```

```python
p3[0::2, 0] # for rows
```

```python
p3[0::2, 0, ::2] # for columns
# Here we take (0::2) first adn last column , so we did jump using this, and we 
# took (0) for first row , and we (::2) ignored middle column
```
