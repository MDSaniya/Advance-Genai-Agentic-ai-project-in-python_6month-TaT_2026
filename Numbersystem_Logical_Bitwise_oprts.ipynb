## Swapping Variables
#### Swappig means exchanging values


#### Using Third Variable
##### temp = a
##### a = b
##### b = temp

```python
a = 5
b = 6
temp = a
a = b
b = temp

a
```

### Without Third Variable
#### a = a+b
#### b = a - b
#### a = a - b

```python
a = 5
b = 6
a = a+b
b = a-b
a = a-b
b
```

# XOR - Exclusive OR - ^
### Differnt bits - 1
### Same bits - 0

```python
10 ^ 5
```

## XOR Swapping
### no extra variable needed

```python
a = 5
b = 6
a = a ^ b
b = a ^ b
a = a ^ b
a
```

# BITWISE OPERATORS
## Bitwise AND
### both must be 1

```python
12 & 13
```

## Bitwise OR |
### At least one bit should be 1

```python
12 | 13
```

## Bitwise XOR
### Different bits - 1

```python
25 ^ 30
```

## Complement ~
### Reverse bits

```python
~12
```

## LEFT SHIFT <<
### Shifts bits to left

```python
10 << 2
```

## RIGHT SHIFT >>
### Shifts bits to right

```python
20 >> 2
```
