# SET OPERATION

## UNION

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {8,9,10}
```

```python
A | B #Union of A and B (All elements from both sets.No Duplicates)
```

```python
A.union(B) # Union of A and B
```

```python
A.union(B,C) # Union of A,B,and C
```

```python
"""Updtes the set callng the update() method with union of A,B and C.
for below example set A will be Updated with union of A,B and C.
"""
A.update(B,C)
A
```

## Intersection

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A & B # Intersection of A and B (common items in both sets)
```

```python
A.intersection(B) Intersection of A and B
```

```python
A.intersection_update(B)
A
```

# Difference

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A - B #set of elements that are only in A but not in B
```

```python
A.difference(B) 
```

```python
B - A
```

```python
B.difference(A)
```

```python
B.difference_update(A)
B
```

## Symmetric Difference

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A ^ B # Symmetric diff (set of elements in A and B but not in both
```

```python
A.symmetric_difference(B) # Symmetric diff of sets
```

```python
""" 
Updates the set calling the symmetric_difference_update() method with the symmet 
For below example Set A will be updated with the symmetric difference of A & B. 
""" 
A.symmetric_difference_update(B)
A
```
